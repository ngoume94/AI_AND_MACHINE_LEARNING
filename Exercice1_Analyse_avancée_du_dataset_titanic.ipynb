```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# charger le dataset
df = sns.load_dataset("titanic")
```

```python
df
```

```python
# 1 - Affichez un countplot des survivants en fonction du sexe et de la classe (hue="sex", x="class").

survivors = df[df['survived'] ==1]

sns.countplot(data= survivors, x = 'class', hue='sex' )
plt.show()
```

```python
# 2 - Affichez un violinplot du prix du ticket par classe (fare vs class).

plt.figure(figsize=(10,5))
sns.violinplot(data=df, x = 'class', y = 'fare')
plt.show()
```

```python
# 3 - Affichez une heatmap des corrÃ©lations entre les variables numÃ©riques.

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Matrice de corrÃ©lation")
plt.show()
```
