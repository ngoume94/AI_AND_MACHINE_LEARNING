```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
```

```python
df
```

```python
# 1 - Convertissez le dataset en pivot table avec year en ligne, month en colonne, values='passengers'.

pivot_df = df.pivot_table(index=['year'], columns=['month'], values=['passengers'])
pivot_df
```

```python
# 2 - Affichez une heatmap de cette table

plt.figure(figsize=(20, 12))
sns.heatmap(pivot_df, annot=True, cmap='coolwarm')
plt.xlabel('X_Axis')
plt.ylabel('Y_Axis')
plt.show()
```

```python
# 3 - Tracez un lineplot des passagers par mois en regroupant par annÃ©e (courbe par annÃ©e)

plt.figure(figsize=(10,5))
sns.lineplot(data=df, x = 'month', y = 'passengers', hue='year')
plt.show()
```
