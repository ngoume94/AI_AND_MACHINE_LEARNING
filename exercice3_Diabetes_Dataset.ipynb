```python
import pandas as pd
import seaborn as sns
```

```python
# charger le dataset
df = pd.read_csv('diabetes_dataset__2019.csv')
df
```

```python
df.describe()
```

```python
# identifier les colonnes avec les valuers manquantes
df.isnull().sum()
```

```python
# Utiliser l'imputation par la médiane pour les variables numériques.
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# Utiliser l'imputation par le mode pour les variables catégorielles.
df['Pdiabetes'].fillna(df['Pdiabetes'].mode()[0], inplace=True)
```

```python
df['Diabetic'].fillna(df['Diabetic'].mode()[0], inplace=True)
```

```python
# verifier s'il existe encore des valeurs manquantes
df.isnull().sum()
```
