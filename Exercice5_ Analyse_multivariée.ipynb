```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
# 1 - un lmplot() pour visualiser la relation entre total_bill et tip, segmentÃ© par sex et smoker.

sns.lmplot(data =df, x='total_bill', y='tip', hue='sex', col='smoker')
plt.show()
```

```python
# 2 - un boxenplot de tip par day et sex

sns.boxenplot(data=df, x='day', y='tip', hue='sex')
plt.show()
```

```python
# 3 - un facetgrid pour afficher des histplot de tip par time et smoker.

g = sns.FacetGrid(df, col='smoker', hue='time')
g.map_dataframe(sns.histplot, x='time', y='tip')
plt.show()
```
