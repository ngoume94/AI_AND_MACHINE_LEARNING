```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# charger le dataset
df = sns.load_dataset('penguins')
```

```python
df
```

```python
# 1 - Affichez un pairplot entre les dimensions (longueur/largeur du bec, taille des nageoires) avec hue="species".

plt.figure(figsize=(12, 8))
sns.pairplot(df,
             x_vars=['bill_depth_mm', 'bill_length_mm', 'flipper_length_mm'], hue='species')
plt.show()
```

```python
# 2 - Utilisez un boxplot pour comparer la masse corporelle par Ã®le.

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, y = 'body_mass_g', x  = 'island')
plt.show()
```

```python
# 3 - Utilisez un scatterplot avec style="sex" pour voir si le sexe influence les mesures.

plt.figure(figsize=(10,5))
sns.scatterplot(data=df, y='sex', x ='flipper_length_mm', style='sex')
plt.show()
```
